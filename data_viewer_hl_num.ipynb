```python
from pathlib import Path
from matplotlib import pyplot as plt
import numpy as np

folder_pth = Path(f"results/hl_num_study")
csv_files = list(folder_pth.glob("*.csv"))
print(type(csv_files))

def extract_sorting_key(path):
    base = path.stem.split('_')[0]
    decimal_part = base.split('.')[1][:4]
    return int(decimal_part)

epoch = []
loss = []
sorted_files = sorted(csv_files, key=extract_sorting_key)

for files in csv_files:
    #print(files)
    data = np.loadtxt(files, delimiter=",")
    epoch.append(data[:,0])
    loss.append(data[:,1])
```

```python
i_range = 3
j_range = 3

fig, axs = plt.subplots(i_range,j_range, figsize=(20, 14), sharey="all")
for i in range(i_range):
    for j in range(j_range):
        index = i*j_range + j
        #print(index)
        #lr = 0.025 + index * 0.050
        #print(lr)
        axs[i,j].plot(epoch[index],loss[index], 'r-')
        axs[i,j].set_xlabel(f"index: {index}")
fig.supxlabel("loss", x=0.5, y=0.05)
fig.supylabel("epoch", x=0.08, y=0.5)
fig.suptitle("hidden layer number vs loss/epoch", x=0.5, y=0.95)
```

```python
import matplotlib.cm as cm
import matplotlib.colors as mcolors

i_range = 3
j_range = 3


number_o_lines = j_range
fig2, ax2 = plt.subplots(i_range, figsize=(12,8),sharey="all")
cmap = cm.get_cmap('Purples', number_o_lines)
start, stop = 0.4, 1.0
colors = [cmap(start + (stop - start) * i / (number_o_lines - 1)) for i in range(number_o_lines)]

for i in range(i_range):
    for j in range(j_range):
        index = i*j_range + j
        #print(index)
        #print(lr)
        if j == 0:
            ax2[i].plot(epoch[index],loss[index], 'o', label=f"HLN: {index}", color=colors[j]) 
        elif j == 1:
            ax2[i].plot(epoch[index],loss[index], 'o', label=f"HLN: {index}", color=colors[j])
        elif j == 2:
            ax2[i].plot(epoch[index],loss[index], 'o', label=f"HLN: {index}", color=colors[j])
            

ax2[0].legend(title="HLN Rate", loc=7)
ax2[1].legend(title="HLN Rate", loc=7)
ax2[2].legend(title="HLN", loc=7)
fig2.supxlabel("epoch", x=0.5, y=0.05)
fig2.supylabel("loss", x=0.08, y=0.5)
fig2.suptitle("HLN vs loss/epoch", x=0.5, y=0.95)
```
